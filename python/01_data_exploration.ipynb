```python
import pandas as pd

# Load dataset
df = pd.read_csv("../data/insurance.csv")

# Show first 5 rows
df.head()
```

```python
df.info()

df.describe()

df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['charges'], bins=30, kde=True)
plt.show()

sns.histplot(df['age'], bins=20, kde=True)
plt.show()
```

```python
bins = [0, 18, 30, 45, 60, 100]
labels = ['0-18', '19-30', '31-45', '46-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df['sex_numeric'] = df['sex'].map({'male':0, 'female':1})
df['smoker_numeric'] = df['smoker'].map({'no':0, 'yes':1})
```

```python
df['cost_per_child'] = df['charges'] / (df['children'] + 1)
```

```python
numeric_cols = ['age', 'bmi', 'children', 'charges', 'sex_numeric', 'smoker_numeric', 'cost_per_child']

plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.show()
```
